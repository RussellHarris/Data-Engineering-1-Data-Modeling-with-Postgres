```python
%sql SELECT 'songs' as table_name, count(*) from songs \
     UNION ALL \
     SELECT 'artists' as table_name, count(*) from artists \
     UNION ALL \
     SELECT 'users' as table_name, count(*) from users \
     UNION ALL \
     SELECT 'songplays' as table_name, count(*) from songplays \
     UNION ALL \
     SELECT 'time' as table_name, count(*) from time;
```

```python
%sql SELECT * \
       FROM songplays \
      WHERE song_id <> 'None' \
        AND artist_id <> 'None' \
      LIMIT 5;
```

```python
%sql SELECT * FROM songplays LIMIT 5;
```

```python
%sql SELECT * \
       FROM songplays \
      WHERE session_id IS NULL OR session_id < 0 OR session_id = 0 \
            OR location IS NULL OR trim(location) = '' \
            OR user_agent IS NULL OR trim(user_agent) = '';
```

```python
%sql SELECT * FROM users LIMIT 5;
```

```python
%sql SELECT * \
       FROM users \
      WHERE first_name IS NULL OR trim(first_name) = '' \
            OR last_name IS NULL OR trim(last_name) = '' \
            OR gender IS NULL OR trim(gender) = '' \
            OR level IS NULL OR trim(level) = '' ;
```

```python
%sql SELECT * FROM songs LIMIT 5;
```

```python
%sql SELECT * \
       FROM songs \
      WHERE title IS NULL OR trim(title) = '' \
            OR artist_id IS NULL OR trim(artist_id) = '' \
            OR year IS NULL OR year < 0 OR year = 0 \
            OR duration IS NULL OR duration = 'NAN';
```

```python
%sql SELECT * FROM artists LIMIT 5;
```

```python
%sql SELECT * \
       FROM artists \
      WHERE name IS NULL  OR trim(name) = '' \
            OR location IS NULL OR trim(location) = '' \
            OR lattitude IS NULL \
            OR longitude IS NULL --or longitude = 'NAN' \
      LIMIT 5;
```
